## [Bayes Rule, 5 points]

After your yearly checkup, the doctor has bad news and good news. The
bad news is that you tested positive for a very serious cancer and
that the test is 99% accurate i.e. the probability of testing
positive given you have the disease is 0.99. The probability of
testing negative if you donâ€™t have the disease is the same. The good news is that it is a very rare condition affecting only 1 in 10,000 people. What is the probability you actually have the disease? (Show all calculations and the final result.)

### Answer

Let $A$ be the event of having the disease, and $B$ be the event of the test being positive. We now define the following probabilities:

* $ P(A) = 1/10000 = 0.0001 $ is the probability of having the disease (this has already been given to us).
* $ P(B) $ is the probability of the test being positive (this will be calculated down below).
* $ P(A|B) $ is the probability of having the disease given that the test is positive (this is what we want to find).
* $ P(B|A) = 0.99 $ is the probability of the test being positive given that you have the disease (this is also given).
* $ P(\bar{B}|\bar{A}) = 0.99 $ is the probability of the test being negative given that you don't have the disease (this is also given).

<br>By conditional probability, the probability of having the disease **and** the test being positive is:
<center>$ P(A \cap B) = P(B|A) \times P(A) = 0.99 \times 0.0001 = 0.000099 $</center>

<br>Also by conditional probability, the probability of **not** having the disease **and** the test being positive is:
<center>$ P(\bar{A} \cap B) = P(B|\bar{A}) \times P(\bar{A}) = (1 - P(\bar{B}|\bar{A})) \times (1 - P(A)) = (1 - 0.99) \times (1 - 0.0001) = 0.01 \times 0.9999 = 0.009999 $</center>

<br>By a simple use of Venn's diagram, we can calculate $P(B)$ as follow:
<center>$ P(B) = P(A \cap B) + P(\bar{A} \cap B) = 0.000099 + 0.009999 = 0.010098 $</center>

<br>Bayes' Theorem can now be applied to compute $P(A|B)$:
<center>$ P(A|B) = \dfrac{P(B|A) \times P(A)}{P(B)} = \dfrac{0.99 \times 0.0001}{0.010098} \approx 0.0098 $</center>


## [Correlation and Independence, 5 points]

Let $X$ be a continuous variable, uniformly distributed in $[-1, +1]$ and let $Y := X^2$. Clearly $Y$ is not independent of $X$ -- in fact it is uniquely determined by $X$. However, show that $\mbox{cov}(X, Y ) = 0$.


### Answer

The probability density function for a continuous uniform distribution is:
<center>$ f(x) = \begin{cases} 1/(b-a) &\mbox{if } a \leq x \leq b \\
                               0       &\mbox{otherwise}\end{cases} $ </center>

Since the given limits are $a = -1$ and $b = 1$, the probability density function for $X$ is simply $1/2$. The expected value of $X^k$ (for any integer $k > 0$) can be calculated using simple integral calculus:
<center>$ E(X^k) = \displaystyle\int\limits_{-1}^{1}x^k f(x)\;\mathrm{d}x
                 = \dfrac{1}{2}\displaystyle\int\limits_{-1}^{1}x^k\;\mathrm{d}x
                 = \dfrac{x^{k+1}}{2(k+1)}\Biggr|_{-1}^{+1}
                 = \dfrac{1-(-1)^{k+1}}{2(k+1)}
$</center>

Using the formula derived, we have:
<center>$ \begin{matrix} E(X^1) = \dfrac{1-(-1)^{2}}{2 \times 2} = 0 &&&
                         E(X^2) = \dfrac{1-(-1)^{3}}{2 \times 3} = \dfrac{1}{3} &&&
                         E(X^3) = \dfrac{1-(-1)^{4}}{2 \times 4} = 0 \end{matrix}
$</center>

We now use the definition of covariance (see [here](http://www.stat.yale.edu/~pollard/Courses/241.fall97/Variance.pdf "Covariance notes from Yale")) and linearity of expectations to calculate the covariance as follows: 
<center>$ \operatorname{cov}(X,Y) = \operatorname{cov}(X,X^2) 
                                  = E((X-E(X))(X^2-E(X^2)))
                                  = E\bigg(X\bigg(X^2-\dfrac{1}{3}\bigg)\bigg) 
                                  = E\bigg(X^3 - \dfrac{1}{3}X\bigg)
                                  = E(X^3) - \dfrac{1}{3}E(X) = 0
$</center>

## [Setting hyperparameters, 3 points]

Suppose $\theta \sim \mbox{Beta}(a,b)$ and we believe $E[\theta] = m$
and $\mbox{var}(\theta) = v$. How should the parameters $a$ and $b$ be
set to be consistent with this?

### Answer

From this [article](https://en.wikipedia.org/wiki/Beta_distribution "Beta distribution Wikipedia"), the expected value of $\theta$ is:
<center>$ \bigg(m = \dfrac{a}{a+b}\bigg) \Leftrightarrow 
          \bigg(a+b = \dfrac{a}{m}\bigg) \Leftrightarrow 
          \bigg(b = \dfrac{a(1-m)}{m}\bigg) 
$</center>

And the variance is: 
<center>$ v = \dfrac{ab}{(a+b)^2(a+b+1)} $</center>

By replacing $(a+b)$ with $(a/m)$, and $(b)$ with $(a(1-m)/m)$, we get:
<center>$ v = \dfrac{a^2(1-m)}{m(a^2/m^2)((a+m)/m)} = \dfrac{m^2(1-m)}{a+m} $</center>

Solving for $a$:
<center>$ a = \dfrac{m^2(1-m)-mv}{v} = \dfrac{m(m-m^2-v)}{v} $</center>

Solving for $b$:
<center>$ b = \dfrac{a(1-m)}{m} = \bigg(\dfrac{m(m-m^2-v)}{v}\bigg) \bigg(\dfrac{1-m}{m}\bigg) = \dfrac{(m-m^2-v)(1-m)}{v} $</center>

## [Plotting normal distributed points, 5 points]

Generate $1000$ points from 2D multivariate normal
distribution having mean $\mu = \left[
\begin{array}{c}
  1  \\
  1
\end{array}
\right]$ and covariance $\Sigma =
\left[
  \begin{array}{rr}
    0.1 & -0.05 \\
    -0.05&  0.2
  \end{array}
\right]
$.  Define the function $f({\bf x}, r) := \frac{({\bf x} - \mu)^{ \top } * \Sigma^{-1} *
  ({\bf x} - \mu) }{ 2} - r $. On a single plot, show the following:
* The level sets $f({\bf x}, r) = 0$ for $r=1, 2, 3$.
* Scatter plot of randomly generated points with points lying
outside $f({\bf x} , 3) = 0$ showing in black while points inside shown in
blue.
* Title of the plot showing how many points lie outside $f({\bf
    x}, 3) = 0$.
Submit your final plot as well as your implementation.



### Discussion

The vectorized version of the method that calculates the f-values gives a huge performance boost. We originally wrote a script that computed the f-values using some nested for-loops, which was slower than our current implementation by approximately 40 times (tested with the `timeit` module). Let the inverse of the covariance and $ \mathbf{x} - \mu $ be:
<center>$ 
  \begin{matrix} \Sigma^{-1} = \begin{bmatrix} a_{00} & a_{01} \\ a_{10} & a_{11} \end{bmatrix} &&&
                 \mathbf{x} - \mu = \begin{bmatrix} x \\ y \end{bmatrix} \end{matrix} 
$</center>

Then clearly, the numerator of the f-value formula can be calculated as below (and this is exactly what our code does):
<center>$ (\mathbf{x} - \mu)^\mathsf{T} \Sigma^{-1}(\mathbf{x} - \mu) 
          = x^2a_{00} + xy(a_{01} + a_{10}) + y^2a_{11} 
$</center>

We create a mesh grid and compute the f-values for each point on the grid. But this is purely for the purpose of drawing the contours for the three level sets. We also compute the f-values for our randomly generated dataset. These points are scattered on the same graph, where points that are inside the outermost contour are colored blue, and points that are ouside are colored black.

```python
import numpy as np
import matplotlib.pyplot as plt

### Vectorized version of the f-value computations.
### OBS: X,Y must be of the same shape. The result Z will be of the same shape.
def f(X, Y, xmean, ymean, cov, r):
  covI = np.linalg.inv(cov)
  XX = X - xmean
  YY = Y - ymean
  Z = XX*XX * covI[0][0]
  Z += YY*YY * covI[1][1]
  Z += XX*YY * (covI[0][1] + covI[1][0])
  return Z / 2.0 - r

### Define the mean and the covariance.
cov = [[0.1, -0.05], [-0.05, 0.2]]
mean = [1, 1]
xmean, ymean = mean

### Create 1000 points normally distributed.
data = np.random.multivariate_normal(mean, cov, 1000)
xmin, ymin = np.amin(data, axis=0)
xmax, ymax = np.amax(data, axis=0)
dataX = data[:,0]
dataY = data[:,1]

### Create a mesh grid
xlist = np.linspace(xmin, xmax, num=1000)
ylist = np.linspace(ymin, ymax, num=1000)
gridX, gridY = np.meshgrid(xlist, ylist)

### Compute all the f-values
gridZ1 = f(gridX, gridY, xmean, ymean, cov, 1) # r = 1
gridZ2 = f(gridX, gridY, xmean, ymean, cov, 2) # r = 2 
gridZ3 = f(gridX, gridY, xmean, ymean, cov, 3) # r = 3
dataZ3 = f(dataX, dataY, xmean, ymean, cov, 3) # r = 3

### Compute the points inside and outside the level set of r = 3
colour = lambda z: 'k' if z > 0 else ('b' if z < 0 else 'r')
colours = [colour(z) for z in dataZ3]
outside = list(filter(lambda x: x > 0, dataZ3))

### Plotting
plt.xlabel('x')
plt.ylabel('y')
plt.title("Number of points outside is " + str(len(outside)))
plt.contour(gridX, gridY, gridZ1, 1, linewidths=2, colors="r")
plt.contour(gridX, gridY, gridZ2, 1, linewidths=2, colors="r")
plt.contour(gridX, gridY, gridZ3, 1, linewidths=2, colors="r")
plt.scatter(dataX, dataY, c=colours, s=20)
plt.show()
```

### Discussion

The correlation color maps of $X$ and $Y$ are exactly the same because the correlation between any two features would not change when we normalize the dataset by scaling the values of each feature between $[0,1]$.

According to the scatter plot, the features with indices 7 and 11 are the ones that have the least correlation. This agrees with the color map for the correlation of $Y$ (because the cell $(7,11)$ has the most reddish tone compared to all the other cells on the map). The correlation for these 2 features is only 0.038, explaining why the scatter plot looks so "random".

Covariance matrices, unlike correlations, cannot be used to unpack the relationship between 2 features in the dataset. This is because covariance is like an unstandardized version of correlation, and what this means is that covariance is measured in the units of the original datasets. Since $Y$ is the normalized version of $X$, the color bar for the covariance of $Y$ is much smaller in scale compared to the color bar for the covariance of $X$. Because this reason, we don't need to pay attention to the colors on the covariance maps. 


```python
import numpy as np
import matplotlib.pyplot as plt

### Uncomment to download the file containing the data
#!pip install -q wget
#import wget
#wget.download("http://www.cse.chalmers.se/research/lab/mlcourse/hw0/dataset0.txt")

### Function to plot a color map
def plotColormap(M, title):
  xlim, ylim = M.shape
  plt.pcolor(M, cmap="RdBu")
  plt.xticks(np.arange(0.5, xlim), range(xlim))
  plt.yticks(np.arange(0.5, ylim), range(ylim))
  plt.title(title)
  plt.colorbar()
  plt.show()

### Import the data and compute the covariance and correlation for X
X = np.loadtxt("./dataset0.txt")
XCov = np.cov(X, rowvar=False)
XCor = np.corrcoef(X, rowvar=False)

### Normalize each feature, and compute the covariance and correlation for Y
Y = X / np.amax(X, axis=0)
YCov = np.cov(Y, rowvar=False)
YCor = np.corrcoef(Y, rowvar=False)

### Plotting color maps
plotColormap(XCov, "Covariance of X")
plotColormap(YCov, "Covariance of Y")
plotColormap(XCor, "Correlation of X")
plotColormap(XCor, "Correlation of Y")

### Scatter plot
i,j = np.unravel_index(np.argmin(YCor), YCor.shape)
title = "Least Correlated Features\nIndices (" + str(i) + "," \
        + str(j) + "), Correlation (" + str(YCor[i][j]) + ")"
plt.scatter(Y[:,i], Y[:,j], c="k", s=30)
plt.title(title)
plt.show()
```
